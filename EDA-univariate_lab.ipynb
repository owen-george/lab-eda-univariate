```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv("amz_uk_price_prediction_dataset.csv")
```

```python
#Generate a frequency table for the product category
freq_table = df['category'].value_counts()
```

```python
#Which are the top 5 most listed product categories?
freq_table.sort_values(ascending=False).head(5)
```

```python
# Display the distribution of products across different categories using a bar chart.
# If you face problems understanding the chart, do it for a subset of top categories.

top_freq_table = freq_table.head(10)

ax = sns.barplot(x=top_freq_table.index, y=top_freq_table.values);
plt.xticks(rotation=45, fontsize=6, ha="right")
plt.xlabel("Top Categories", fontsize=10);
```

```python
# Part 2: Delving into Product Pricing
# Business Question: How are products priced on Amazon UK, and are there specific price points or ranges that are more common?

#Measures of Centrality:
# Calculate the mean, median, and mode for the price of products.
df['price'].describe().round(2)
```

```python
df['price'].mode()[0]
```

```python
# What's the average price point of products listed? How does this compare with the most common price point (mode)?

# The average price of 89.24 is well above the median and mode, skewed upwards by some very high value products (up to 100,000, whereas 75% are below 50)
```

```python
#Measures of Dispersion:
#Determine the variance, standard deviation, range, and interquartile range for product price.

var_price = df['price'].var()
std_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()
iqr_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)

var_price, std_price, range_price, iqr_price
```

```python
# How varied are the product prices? Are there any indicators of a significant spread in prices?

# There is a massive range in prices - The std is larger than the mean and interquartile range
```

```python
#Visualizations:
#Is there a specific price range where most products fall?
# Plot a histogram to visualize the distribution of product prices.
# If its hard to read these diagrams, think why this is, and explain how it could be solved..

sns.histplot(df['price'], kde=True, bins=50, color="green")

# Hard to read since the maximum value is so much greater than where the vast majority of values lie
```

```python
price_bins = [0, 1, 10, 50, 100, 200, 500, 1000, 5000, df['price'].max()]
labels = ['0-1', '1-10', '10-50', '50-100', '100-200', '200-500', '500-1000', '1000-5000', '5000+']
df['price_groups'] = pd.cut(df['price'], bins=price_bins, labels=labels, include_lowest=True)

sns.countplot(data=df, x='price_groups', palette='Set2', hue='price_groups', legend=False,
              order=['0-1', '1-10', '10-50', '50-100', '100-200', '200-500', '500-1000', '1000-5000', '5000+']);
plt.xticks(rotation=45, fontsize=10, ha="right");
```

```python
# Are there products that are priced significantly higher than the rest?
#Use a box plot to showcase the spread and potential outliers in product pricing.

sns.boxplot(x = df['price'], color="lightblue")
```

```python
# Part 3: Unpacking Product Ratings
# Business Question: How do customers rate products on Amazon UK, and are there any patterns or tendencies in the ratings?

# Measures of Centrality:
# Calculate the mean, median, and mode for the rating of products.
df_reviewed = df[df['reviews']>0]

mean_stars = df_reviewed['stars'].mean()
median_stars = df_reviewed['stars'].median()
mode_stars = df_reviewed['stars'].mode()[0]

mean_stars, median_stars, mode_stars
```

```python
# How do customers generally rate products? Is there a common trend?

#Most products have ratings around 4.5 stars, there isn't much variation from the mean/mode/median
```

```python
# Determine the variance, standard deviation, and interquartile range for product rating.

var_stars = df_reviewed['stars'].var()
std_stars = df_reviewed['stars'].std()
range_stars = df_reviewed['stars'].max() - df_reviewed['stars'].min()
iqr_stars = df_reviewed['stars'].quantile(0.75) - df_reviewed['stars'].quantile(0.25)

var_stars, std_stars, range_stars, iqr_stars
```

```python
# Are the ratings consistent, or is there a wide variation in customer feedback?

# Ratings are largely consistent, without much variation (most are 4-5 stars)
```

```python
# Calculate the skewness and kurtosis for the rating column.

df_reviewed['stars'].skew()
```

```python
df_reviewed['stars'].kurtosis()
```

```python
# Are the ratings normally distributed, or do they lean towards higher or lower values?

# ratings are negatively skewed by a moderate amount 
```
